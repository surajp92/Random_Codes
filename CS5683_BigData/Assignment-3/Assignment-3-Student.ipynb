##### Question-5: Do you think we can achieve PCA with SVD? If so, how? (2 Points)

Yes, we can do PCA with SVD results also. In case of the PCA, first we form the correlation matrix, i.e., **$M^T M$** and then compute the principal direction or the eigenvectors.  The right singualr vectors are basically the principal component or eigenvectors of the matrix **$M^T M$**. For example, using SVD, we have **$M = U \Sigma V^T$**, and **$M^T M = V \Sigma U^T U \Sigma V^T = V \Sigma^2 V^T$**. Therefore, **$V$** is the eigenvector of **$M^T M$** and the square of singualar values are the eigenvalues of the correlation matrix **$M^T M$**.
